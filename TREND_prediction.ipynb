# Part1 - Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load data

```python
import os
os.chdir('/Users/shiliu/Python/Projects/essay/trend')
file_chdir = os.getcwd()

filecsv_list = []
for root, dirs, files in os.walk(file_chdir):
    for file in files:
        if os.path.splitext(file)[1] == '.csv':
            filecsv_list.append(file)
filecsv_list
```

```python
trend = pd.read_csv('trend.csv')
dji = pd.read_csv('DJI_monthly.csv')
target = dji.iloc[:, -1]

dataset = pd.concat([trend, target],axis=1).dropna()
dataset = dataset.iloc[:, 1:]
dataset
```

Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
dataset = sc.fit_transform(dataset)
pd.DataFrame(dataset)
```

```python
X = pd.DataFrame(dataset).iloc[:, 0:-1]
y = pd.DataFrame(dataset).iloc[:, -1]

X
```

Split the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                     test_size = 0.1, 
                                                     random_state = 0)
#test_size = 0.2 = 20%
```

```python
pd.DataFrame(X_train)
```

Reshape - 3 dimensions

```python
X_train = np.reshape(X_train, (X_train.shape[0], 
                               X_train.shape[1]))
```

# Part2 - Building the RNN

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

Initializing the RNN

```python
regressor = Sequential()
```

Adding the first LSTM layer and also some Dropout regularisation

```python
regressor.add(LSTM(units = 50, 
                   return_sequences = True, 
                   input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))
```

Adding the second LSTM layer and also some Dropout regularisation

```python
regressor.add(LSTM(units = 50, 
                   return_sequences = True))
regressor.add(Dropout(0.2))
```

Adding the third LSTM layer and also some Dropout regularisation

```python
regressor.add(LSTM(units = 50, 
                   return_sequences = True))
regressor.add(Dropout(0.2))
```

Adding the fourth LSTM layer and also some Dropout regularisation

```python
regressor.add(LSTM(units = 50, 
                   return_sequences = True))
regressor.add(Dropout(0.2))
```

Adding the output layer

```python
regressor.add(Dense(units = 1))
```

Compiling the RNN

```python
regressor.compile(optimizer = 'adam',
                  loss = 'mean_squared_error')
```

Fitting thr RNN to the training set

```python
regressor.fit(X_train, y_train, epochs = 100, batch_size = 32, verbose =1)
```
